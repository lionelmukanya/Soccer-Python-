```python
#How many unique referees are there?
df['Referee'].unique()
```

```python
#Which referees whistled fo the most fouls?
df['Fouls']= df['HF'] + df['AF']
df.groupby('Referee')['Fouls'].sum().sort_values(ascending=False).head(5)


#Referee A Taylor called the most fouls
```

```python
#Pie chart of Fouls called by all the referees

a = df.groupby(['Referee']).sum()['Fouls']
labels = ['A Madley','A Marriner','A Taylor','C Kavanagh','C Pawson','D Coote',
'G Scott','J Moss','K Friend','L Mason','L Probert','M Atkinson','M Dean','M Oliver',
'P Tierney','R East','S Attwell',
'S Hooper']
fig, ax = plt.subplots()
ax.pie(a,  labels=labels, autopct='%1.1f%%', shadow=False, startangle=90)
ax.axis('equal')
ax.set_title("Fouls Call Share by referees")
plt.show()

```

```python
#Grouping the Data by Full time Result
df.groupby(['FTR']).sum()
```

```python
#Bar graph that groups the Foul count by full time result

a = df.groupby(['HTR']).sum()['Fouls']
labels = ['Away Win', 'Draw', 'Home Win']

fig, ax = plt.subplots()
ax.bar(range(len(a)),a, label= labels, color = [ "brown", "magenta", "beige"])

plt.xlabel("Full Time Result")
plt.ylabel("Fouls")
plt.title("Foul Count in each Full Time Result")
plt.xticks(range(len(a)), labels)

plt.show()

#Conclusion: There more fouls in matches that end in a draw.
```

```python
#Is there a correlation between goals and corners taken?

df['goals'] = df['FTHG'] + df['FTAG']

df['corners'] = df['HC'] + df['AC']


df[['goals', 'corners']]

fig, ax = plt.subplots()

# Plot data on the axes
x = df['goals']
y = df['corners']
ax.scatter(x, y, color = 'purple')

# Set labels and title
ax.set_xlabel('Goals')
ax.set_ylabel('Corners')
ax.set_title('Goals vs Corners')

# Show the plot
plt.show()

#No correlations
```

```python
#Is there a correlation between fouls and cards (red and Yellow)?

df['cards'] = df['AR'] + df['HR'] + df['AY'] + df['HY']

df['goals'] = df['FTHG'] + df['FTAG']
fig, ax = plt.subplots()

# Plot data on the axes
x = df['cards']
y = df['goals']
ax.scatter(x, y, color = 'crimson')

# Set labels and title
ax.set_xlabel('Goals')
ax.set_ylabel('Cards')
ax.set_title('Goals vs Cards')

plt.show()

#No Correlation
```

```python
#Which home team had the most wins?

#How many unique Home teams are there?
df['HomeTeam'].unique()


b = df.loc[df['FTR']=='H']

b.groupby(['HomeTeam']).count()['FTR'].sort_values(ascending=False).head(5)

#Man City had the most home wins.
```

```python
#Which team had the most away draws?
df['AwayTeam'].unique()

c = df.loc[df['FTR'] == 'D']

c.groupby(['AwayTeam']).count()['FTR'].sort_values(ascending=False).head(5)

#NewCastle had the most away draws.
```
